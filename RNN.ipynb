```python
import os
import pandas as pd
import numpy as np

#Directory of dataset
data_dir = "data/release/taxi_log_2008_by_id"

#Directory of dataset after cleansing
output_dir = "cleaned_data"
os.makedirs(output_dir, exist_ok=True)

#Access all .txt files
all_files = [f for f in os.listdir(data_dir) if f.endswith(".txt")]
print(f"Found {len(all_files)} text files in {data_dir}.")

all_trajectories = []

for filename in all_files:
    file_path = os.path.join(data_dir, filename)
    
    try:
        df = pd.read_csv(file_path, header=None, names=["taxi_id", "time", "lon", "lat"], sep=",",)
    except Exception as e:
        print(f"Error reading {filename}: {e}")
        continue

    #Data cleansing

    #Treansform time format
    df["time"] = pd.to_datetime(df["time"], format="%Y-%m-%d %H:%M:%S", errors="coerce")
    
    #Delete wrong data
    df.dropna(subset=["time", "lon", "lat"], inplace=True)
   
    #Filter outliers of longtitude and latitude
    df = df[(df["lat"] >= -90) & (df["lat"] <= 90)]
    df = df[(df["lon"] >= -180) & (df["lon"] <= 180)]

    #Sort by time
    df.sort_values(by="time", inplace=True)

    if len(df) < 5:
        continue
    all_trajectories.append(df)
    cleaned_path = os.path.join(output_dir, filename.replace(".txt", "_cleaned.csv"))
    df.to_csv(cleaned_path, index=False)
print(f"After cleaning, we have {len(all_trajectories)} valid trajectories in memory.")   





```
